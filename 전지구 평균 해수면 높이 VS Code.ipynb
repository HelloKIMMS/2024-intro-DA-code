### 4. 전지구 해수면 높이 그래프

```python
import csv
import matplotlib.pyplot as plt

f = open('전지구 해수면높이.csv', encoding='utf-8')
data = csv.reader(f)

header = next(data)

years = []
sea_level_mean = [] 

for row in data:
    try:
        year = int(row[0])
        sea_level = float(row[1]) 
        years.append(year)
        sea_level_mean.append(sea_level)
    except ValueError:
        print(f"유효하지 않은 데이터 건너뜀: {row}")

f.close()

plt.figure(figsize=(10, 4))
plt.plot(years, sea_level_mean, color='b', label='해수면 높이 변화')

plt.title('전지구 평균 해수면 높이 변화')
plt.xlabel('년도 (Year)')
plt.ylabel('평균 해수면 높이 (mm)')

plt.grid()
plt.legend(loc='upper left')

plt.show()
```
